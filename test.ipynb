```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
import scipy as sp
from sympy import *
```

```python
def function(x):
    return (np.exp(-x / 10) * np.sin(x))
x_variable = np.linspace(0, 10, 10000)
y_variable = function(x_variable)
plt.plot(x_variable, y_variable)
plt.show(True)
```
