```python
def g(x):
    return np.cos(x)
    
print("original values")
x = np.linspace(0, 10, 1000)
y = g(x)
plt.plot(x, y, linestyle="--")
print("integrate values")
integral_value = np.cumsum(y) * (x[1] - x[0])
#y will be differentiated with x acting as spacing between two values
#its like differentiating, then you plot in the given (or found values)
#in your differentiated equation
plt.plot(x, integral_value)
```
