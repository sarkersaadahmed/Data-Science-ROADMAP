```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
import scipy as sp
from sympy import *
```

```python
def surface_function(x, y):
    return np.exp(-((x**2) + (y**2)))

x_variable = np.arange(-2, 2)
y_variable = np.arange(-2, 2)
x_value, y_value = np.meshgrid(x_variable, y_variable)
z_variable = surface_function(x_value, y_value)
plt.figure(figsize=(10, 6))
plt.contour(z_variable, x_value, y_value, levels=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('surface plot')
plt.show(True)
```
