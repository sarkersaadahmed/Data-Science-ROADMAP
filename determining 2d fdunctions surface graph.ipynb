```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
import scipy as sp
from sympy import *
```

```python
def surface_function(x, y):
    return ((x**2)*(y) + (2*x*y) + 1)

x_variable = np.linspace(0, 100, 1000)
y_variable = np.linspace(0, 100, 1000)
X, Y = np.meshgrid(x_variable, y_variable)
Z = surface_function(X, Y)

plt.figure(figsize=(10, 6))
plt.contour(X, Y, Z, levels=100)  # Contour plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot of Surface Function')
plt.colorbar(label='Z')
plt.show()
```
